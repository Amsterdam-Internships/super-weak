```python
import csv
import spacy
```

```python
trainpath = "data/ned.train"
```

```python
lines = []

with open(trainpath, 'r', encoding = "cp850") as file:
    reader = csv.reader(file, delimiter = ' ')
    for item in reader:
        if item:
            lines.append(item)
```

```python
counter = 0
container = []
text = []
while counter < len(lines):
    line = lines[counter]
    if line[0] == '-DOCSTART-':
        container.append(text)
        text = []
        counter += 1
    else:
        text.append(line)
        counter += 1
```
